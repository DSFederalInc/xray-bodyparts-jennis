```python
import pandas as pd

df = pd.read_csv('C:/UNIFESP_Xray/Xray_small/train.csv')

print(df.head()) 
```

```python
class_number = len(set(list(df['Target'])))
print("classes number:",len(df['Target'].value_counts()))
```

```python
import os
import shutil

# Move the images into separate folders which named by their labels
for i in range(len(df['Target'])):
    
    path = "destination_path" #the destination you want to move to separated folders
    directory = df['Target'][i]
    save_path = path + directory.replace(" ","_")
    
    if not os.path.exists(save_path):
        os.mkdir(save_path)


    original_path = "C:/UNIFESP_Xray/Xray_small_jpg/train"
    file_name = df['SOPInstanceUID'][i] + "-c.jpg"
    print(file_name)

    shutil.copy(os.path.join(original_path, file_name), os.path.join(save_path, file_name))
```
